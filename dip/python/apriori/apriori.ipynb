```python
from pandas import pandas as pd
movies = pd.read_csv("movies.csv")
ratings = pd.read_csv("ratings.csv")
```

```python
movies.describe(include='all')
```

```python
ratings.describe(include='all')
```

```python
df = pd.merge(movies, ratings, on='movieId')
df.describe(include='all')
```

Tanto df como ratings tienen 100836 rows. con eso compruebo que no perdi datos

filtro pelis mayores a 4

```python
df = df[df['rating']>=4.0]
df.head()
```

```python
apriori_df = df.groupby('userId')['title'].apply(list)
apriori_df
```

```python
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder

te = TransactionEncoder()
te_ary = te.fit(apriori_df).transform(apriori_df)
df2 = pd.DataFrame(te_ary, columns=te.columns_)
df2
```

```python
f=apriori(df2, min_support=0.15, max_len=2, use_colnames=True)
#Esto es una variante a lo que vimos la clase anterior
rules=association_rules(f,metric='lift',min_threshold=2)
```
