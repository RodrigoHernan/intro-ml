```python
import seaborn as sns
import pandas as pd
import numpy as np
df = sns.load_dataset('tips')
df
```

```python
df.describe(include='all')
```

```python
df['day'].value_counts()
```

```python
df['day'].value_counts() / df['day'].value_counts().sum() *  100
```

```python
df.groupby('sex').mean()
```

```python
df['prc_tip'] = df['tip']/df['total_bill']
```

```python
df.groupby('sex').mean()
```

```python
df.groupby('sex')[['total_bill', 'prc_tip']].describe()
```

```python
def mean_aur2usd(x):
    return np.mean(x) * 1.2
```

```python
df.groupby(['sex', 'time'])[['total_bill', 'prc_tip']].apply(mean_aur2usd)
```

```python
df.groupby(['sex', 'time'])[['total_bill', 'prc_tip']].aggregate([mean_aur2usd, np.max])
```

```python
dict_agg = { 
    'prc_tip': [min, max],
    'total_bill': [mean_aur2usd, np.max],
}


df.groupby(['sex', 'time'])[['total_bill', 'prc_tip']].agg(dict_agg)
```

```python
def f_filter(x):
    return mean_aur2usd(x['total_bill'].mean()) > 20

df.groupby(['sex', 'time']).filter(f_filter)
```

```python
df['ones'] = 1
```

```python
df_g = df.groupby(['sex', 'smoker'])[['ones']].sum()
df_g
```

```python
df_g.groupby(level=0).apply(
    lambda x: x/x.sum() * 100
)
```

```python
pd.cut(df['total_bill'], bins=[3, 18, 35, 60]).value_counts()
```

```python
df['bin_total'] = pd.cut(df['total_bill'], bins=[3, 18, 35, 60])
```

```python
df.groupby(['time', 'bin_total'])[['ones']].count()
```

```python
df.groupby(['time', 'bin_total'])[['ones']].count().groupby(level=0).apply(
    lambda x: x/x.sum() * 100
)
```
